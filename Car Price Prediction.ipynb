```python
data.price=data.price.astype(int)
```

#####  creating dummy variables

```python
dummy_fueltype=pd.get_dummies(data.fueltype)
dummy_aspi=pd.get_dummies(data.aspiration)
dummy_carbody=pd.get_dummies(data.carbody,drop_first=True)
dummy_drivewheel= pd.get_dummies(data.drivewheel,drop_first=True)
dummy_enginelocation=pd.get_dummies(data.enginelocation)
dummy_enginetype=pd.get_dummies(data.enginetype,drop_first=True)
dummy_fuelsys=pd.get_dummies(data.fuelsystem,drop_first=True)
dummy_cylinder=pd.get_dummies(data.cylindernumber,drop_first=True)
dummy_doornum=pd.get_dummies(data.doornumber)
```

```python
## concatinating the dummy varuables into the main dataframe

data=pd.concat([data,dummy_fueltype,dummy_aspi,dummy_carbody,dummy_cylinder,dummy_drivewheel,
                dummy_enginelocation,dummy_enginetype,dummy_fuelsys],axis=1)
```

```python
data.head()
```

```python
## Droping cathegorical columns referencing the dummy variables
```

```python
data=data.drop(['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber'],axis=1)
```

```python
df=data.copy()
```

```python
df
```

```python
df=df.drop('company_name',axis=1)
```

```python
df=df.drop('fuelsystem',axis=1)
```

```python
x=df.drop('price',axis=1)
y=df.price
```

## Building the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
x_test.shape
y_train.shape
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
lr.coef_
```

### Calculating the accuracy

```python

print('Training Accuracy:', lr.score(x_train,y_train)*100)
print('Testing Accuracy:', lr.score(x_test,y_test)*100)
```
